```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movie_revenues.csv')
```

```python
df['revenue_mln'] = df['revenue'].apply(lambda x: x / 1000000)
```

```python
def get_mad(s):
    median = np.median(s)
    MAD = np.median(np.abs(s - median))

    return MAD, median
```

```python
MAD, median = get_mad(df['revenue_mln'])
```

```python
df['mod_zscore'] = df['revenue_mln'].apply(lambda x: .6745 * (x - median) / MAD)
```

```python
df[df['mod_zscore'] > 3.5]
```
