```python
import pandas as pd
from sklearn import linear_model
```

```python
data = pd.read_csv('canada_per_capita_income.csv')
```

```python
data.head()
```

```python
data.plot(kind='scatter',x='year',y='per capita income (US$)')
```

```python
lr_model = linear_model.LinearRegression()
lr_model.fit(data[['year']],data[['per capita income (US$)']])
```

```python
lr_model.predict([[2020]])
```
