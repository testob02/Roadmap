```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = plt.imread('serena_williams.jpg')
```

```python
blur = cv2.GaussianBlur(image,(5,5),0)


fig = plt.figure(figsize=(10,6))
ax1 = plt.subplot(1,2,1)
ax1.add_image(plt.imshow(image))
plt.title('Normal Image')
ax2 = plt.subplot(1,2,2)
ax2.add_image(plt.imshow(blur))
plt.title('Gaussian Blur')
plt.show()
```

```python
canny = cv2.Canny(image,100,200)


fig = plt.figure(figsize=(10,6))
ax1 = plt.subplot(1,2,1)
ax1.add_image(plt.imshow(image))
plt.title('Normal Image')
ax2 = plt.subplot(1,2,2)
ax2.add_image(plt.imshow(canny, cmap='gray'))
plt.title('Canny Edge Detection')
plt.show()
```

```python
resize = cv2.resize(image,(200,200))


fig = plt.figure(figsize=(10,6))
ax1 = plt.subplot(1,2,1)
ax1.add_image(plt.imshow(image))
plt.title('Normal Image')
ax2 = plt.subplot(1,2,2)
ax2.add_image(plt.imshow(resize))
plt.title('Resized Image')
plt.show()
```

```python
(h,w) = image.shape[:2]
center = (w//2 , h//2)
rot_mat = cv2.getRotationMatrix2D(center,45,1.0)
rotated = cv2.warpAffine(image,rot_mat,(w, h))


fig = plt.figure(figsize=(10,6))
ax1 = plt.subplot(1,2,1)
ax1.add_image(plt.imshow(image))
plt.title('Normal Image')
ax2 = plt.subplot(1,2,2)
ax2.add_image(plt.imshow(rotated))
plt.title('Rotated Image')
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


fig = plt.figure(figsize=(10,6))
ax1 = plt.subplot(1,2,1)
ax1.add_image(plt.imshow(image))
plt.title('Normal Image')
ax2 = plt.subplot(1,2,2)
ax2.add_image(plt.imshow(gray, cmap='gray'))
plt.title('Grayscale Image')
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
copy = image.copy()
for (x, y, w, h) in faces:
    cv2.rectangle(copy, (x, y), (x+w, y+h), (255, 0, 0), 2)


fig = plt.figure(figsize=(10,6))
ax1 = plt.subplot(1,2,1)
ax1.add_image(plt.imshow(image))
plt.title('Normal Image')
ax2 = plt.subplot(1,2,2)
ax2.add_image(plt.imshow(copy, cmap='gray'))
plt.title('Face Detection')
plt.show()
```
